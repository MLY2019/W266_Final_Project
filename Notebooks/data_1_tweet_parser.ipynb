```python
import os
import tweepy as tw
import pandas as pd
import math
```

```python
# private twitter developer key
consumer_key= 'private_key'
consumer_secret= 'private_secret'

# set up API
auth = tw.OAuthHandler(consumer_key, consumer_secret)
api = tw.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
def get_data_set(file_name):
    '''function to load in data'''

    with open(file_name, 'r') as f:
        contents = f.read()

    contents = contents.split('\n')[:-1] # ignore last row which is empty
    contents = {i.split('\t')[0]: i.split('\t')[2] for i in contents} # dict format: tweet_id: indicator
    return(contents)
```

```python
train_set_1_name = 'Project/rxwfb3tysd-2/subtask1/training_set_1_ids.txt'
train_set_2_name = 'Project/rxwfb3tysd-2/subtask1/training_set_2_ids.txt'

# read in the train tweet_id
train_set_1 = get_data_set(train_set_1_name)
train_set_2 = get_data_set(train_set_2_name)

train_set = {**train_set_1, **train_set_2}
```

```python
# test the API to featch tweet body
tweets = api.statuses_lookup([i for i in train_set.keys()][:100])
```

```python
def get_tweets(tweet_ids, output):
    '''function to loop thru tweet id and get tweet body'''
    length = len(tweet_ids)
    num_iteration = math.ceil(length/100) # each loop will take 100 tweet id
    
    for i in range(num_iteration):
        begining = i * 100
        ending = (i + 1) * 100
        sub_set = tweet_ids[begining:ending]
        
        tweets = api.statuses_lookup(sub_set)
        
        print('{} out of {} tweets added'.format(len(tweets), len(sub_set)))
    
        for tweet in tweets:
           #tweet obtained
           output[tweet.id] = tweet.text
```

```python
# parse train dataset
train_data = {}
get_tweets([i for i in train_set.keys()], train_data)
```

```python
# write train data as text file
train_data_final = ''

for i in train_data:
    train_data_final += '{} \t\t\t\t\t {} \t\t\t\t\t {}\n\n\n\n\n'.format(i, train_data[i], train_set[str(i)])
    
with open('training_data.txt', 'w') as f:
    f.write(train_data_final)
```

```python
# read in test tweet id and parse from twitter

test_set_name = 'Project/rxwfb3tysd-2/subtask1/evaluation_set_ids.txt'

test_set = get_data_set(test_set_name)

test_data = {}
get_tweets([i for i in test_set.keys()], test_data)
```

```python
test_data_final = ''

for i in test_data:
    test_data_final += '{} \t\t\t\t\t {} \t\t\t\t\t {}\n\n\n\n\n'.format(i, test_data[i], test_set[str(i)])
    
    
with open('test_data.txt', 'w') as f:
    f.write(test_data_final)
```
